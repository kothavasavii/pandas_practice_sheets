```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title("simple line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y,color="purple",linestyle="dashed",marker="*",linewidth=1)
```

```python
x=[1,2,3,4,5,6]
y=[2000,350,1000,4000,7000,5000]
plt.plot(x,y,color="green" ,linestyle="dashed", marker="o")
plt.title("savings account over 6 months")
plt.xlabel("6 months")
plt.ylabel("savings amount")
plt.show() 
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=["black","red","green","blue","violet"],edgecolor="purple" )
plt.title("Basic bar chart")
plt.xlabel("categories")
plt.ylabel("Values")
plt.show()
```

```python
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35,36,36,37,38,38,38,39,40,40,41,42,43,44,56,66,70]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color="skyblue",edgecolor="black")
plt.title("Age distribution of customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
lang=["python","Ruby","c++","Java"]
percent=[40.0,10.0,20.0,30.0]
colors=["yellow","pink","grey","skyblue"]
plt.pie(percent,labels=lang,colors=colors, autopct="%1.1f%%" ,startangle=90)
plt.legend()
plt.title("Programming language usage")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.scatter(x,y,color="red" , s=100, edgecolor="black")
plt.title("scatter plot example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```
