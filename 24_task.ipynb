```python
# line plots to track monthly sales 
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [10000, 12000, 15000, 18000, 16000, 17000, 20000, 19000, 22000, 21000, 25000, 27000]
plt.plot(months,sales,color="green",marker="*")
plt.title("Tracking Monthly Sales")
plt.xlabel("Months of Year")
plt.ylabel("Sales")
plt.show() 
```

```python
# histograms for order quantity distribution.
order_quantities = [1, 2, 1, 5, 10, 3, 8, 6, 1, 2, 5, 7, 2, 6, 10, 3, 4, 8, 9, 1, 5, 7]
plt.hist(order_quantities,bins=range(1,11),edgecolor="black")
plt.title("Order Quantity Distribution")
plt.xlabel("Order Quantity (Number of Items)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Customer Demographics:Visualize age distribution(histogram)
import matplotlib.pyplot as plt
import numpy as np
ages = [18, 22, 25, 29, 31, 33, 35, 40, 45, 50, 23, 30, 19, 22, 60, 28, 38, 40, 50, 55, 60]
plt.hist(ages, bins=range(18, 70, 5), edgecolor='black', color='lightblue')
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# spending patterns(scatter plot).
spending = [100, 200, 150, 220, 250, 300, 320, 350, 400, 500, 120, 230, 110, 180, 600, 160, 300, 330, 400, 450, 600]
plt.scatter(ages, spending, color='purple')
plt.title("Spending Patterns by Age")
plt.xlabel("Age")
plt.ylabel("Spending ($)")
plt.tight_layout()
plt.show()
```

```python
# School Performance: Compare students marks across students(bar chart) 
import seaborn as sns
import pandas as pd
students = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Heidi", "Ivan", "Judy", "Karl", "Laura", "Mallory", "Nina", "Oscar",
            "Peggy", "Quincy", "Ralph", "Sybil", "Trent"]
marks = [88, 75, 92, 85, 90, 78, 88, 84, 95, 87,  81, 79, 91, 86, 95, 82, 89, 84, 90, 77]  

df = pd.DataFrame({"Student": students, "Marks": marks})

plt.figure(figsize=(10, 5))
sns.barplot(x="Student", y="Marks", hue="Student", data=df, palette="viridis", legend=False)
plt.xticks(rotation=45)
plt.title("Student Marks Comparison (Bar Chart)")
plt.ylabel("Marks")
plt.show()
```

```python
# identify outliers(box plot).
plt.figure(figsize=(5, 5))
sns.boxplot(y=df["Marks"], palette="pastel")
plt.title("Distribution of Student Marks (Box Plot)")
plt.ylabel("Marks")
plt.show()
```
