```python
import seaborn as sns
tips=sns.load_dataset('tips')
print(tips.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color="skyblue")
plt.title("Distribution of Total bills")
plt.show()
```

```python
sns.kdeplot(data=tips,x="total_bill",color="purple",shade=True)
plt.title("KDE plot of total bills")
plt.show()
```

```python
sns.barplot(data=tips,x="day",y="total_bill",palette="viridis")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",palette="coolwarm")
plt.title("Box plot of Total Bills by Day and Gender")
plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",palette="deep")
plt.title("scatter plot of Total Bill vs Tip")
plt.show()
```

```python
sns.lineplot(data=tips,x="size",y="tip",marker="o", color="blue")
plt.title("Line Plot of party size vs Tip")
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex" ,palette="husl")
plt.show()
```

```python
numeric_tips = tips.select_dtypes(include=['number'])
correlation_matrix=numeric_tips.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
train_data=pd.read_csv("train.csv")
train_data.head()
```

```python
sns.barplot(data=train_data,x="Pclass",y="Survived",errorbar=None,palette=["red","violet","lavender"])
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
sns.scatterplot(data=train_data,x="Age",y="Fare",hue="Survived",palette="coolwarm",alpha=0.7)
plt.title("Age vs Fere")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
selected_data=train_data[["Age","Fare","Pclass","Survived"]]
sns.pairplot(data=selected_data,hue="Survived",palette="coolwarm",diag_kind="kde")
plt.suptitle("Pair wise Relationships",y=1.02)
plt.show()
```

```python
numeric_val = train_data.select_dtypes(include=['number'])
correlation_matrix=numeric_val.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",fmt=".2f",linewidth=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
